```python
# Assumes you have the enron email dataset as emails.csv

import pandas as pd
data = pd.read_csv("enron_05_17_2015_with_labels_fullcsv.zip")
data=data.dropna(subset=['content'],axis=0)
data['content'] = data["content"].apply(lambda val: val.replace("\n",' '))


# Lets look only at emails with 100 words or less and that are Non-replies
data=data.fillna(value={'Subject':''})
df02=data[(data['content'].str.len() <100) & ~(data['Subject'].str.contains('Re:')) & (data['content'].str.match(r'^[a-zA-Z .,]*$'))]
```

```python
import re
def trim_string(s):
    s=re.sub(r'\.+','.',s)
    return re.sub(r'https?:\/\/[0-9A-Za-z.?&=+%\/:_\-]+','',s)
```

```python
string2='\n'.join(df2['content'].apply(lambda x: trim_string(x))).replace('. ','.\n').replace(', ','.\n')
string3='\n'.join([x for x in string2.split('\n') if len(x.split())>1])
```
